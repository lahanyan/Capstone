```python
import pandas as pd

import sys
import sklearn.neighbors._base
sys.modules['sklearn.neighbors.base'] = sklearn.neighbors._base
from missingpy import MissForest

%load_ext autotime
```

```python
filepath = "../Data/TrainDataSample.csv"

train_data = pd.read_csv(filepath)

filepath = "../Data/TestDataSample.csv"

test_data = pd.read_csv(filepath)
```

```python
# Using MissForest to impute missing values of the train dataset.

imputer = MissForest(random_state=42)
train_X_imputed = imputer.fit_transform(train_data)
```

```python
# Using MissForest to impute missing values of the test dataset.

imputer = MissForest(random_state=42)
test_X_imputed = imputer.fit_transform(test_data)
```

```python
# transforming to dataframe scaled columns which are already imputed, replacing main dataframes columns by imputed columns.

scaled_data_for_mf = pd.DataFrame(train_X_imputed, index=train_data.index, columns=train_data.columns)
train_data[scaled_data_for_mf.columns] = scaled_data_for_mf
train_data[scaled_data_for_mf.columns]
```

```python
# transforming to dataframe scaled columns which are already imputed, replacing main dataframes columns by imputed columns.

scaled_data_for_mf = pd.DataFrame(test_X_imputed, index=test_data.index, columns=test_data.columns)
test_data[scaled_data_for_mf.columns] = scaled_data_for_mf
test_data[scaled_data_for_mf.columns]
```

```python
# Concatinating train and test dataframes to check if there are any 0s.

data = pd.concat([train_data, test_data])

Nas = []
count_rows = data.shape[0]
for col in data.columns:
    count_nan = data[col].isna().sum() / count_rows * 100
    Nas.append(count_nan)
NAs_df = pd.DataFrame({'cols':data.columns, 'Nas_perc':Nas})
NAs_df = NAs_df.sort_values(by=['Nas_perc'], ascending=False)
NAs_df
```

```python
# Saving train and test datasets.

train_data.to_csv('../Data/TrainDataMf.csv', index=False) 
test_data.to_csv('../Data/TestDataMf.csv', index=False) 
```
