```python
import pandas as pd
%load_ext autotime
```

```python
filepath = "../Data/TrainDataSample.csv"

train_data = pd.read_csv(filepath)

filepath = "../Data/TestDataSample.csv"

test_data = pd.read_csv(filepath)
```

```python
# Filling missing values by mean if mean is near median and by median otherwise for train dataset's columns.

for i in train_data.columns:
    if(train_data[i].mean() <= (train_data[i].median() * 1.05) or train_data[i].mean() >= (train_data[i].median() * 0.95)):
        train_data[i] = train_data[i].fillna(train_data[i].mean())

    else:
        train_data[i] = train_data[i].fillna(train_data[i].median())
        
        
```

```python
# Filling missing values by mean if mean is near median and by median otherwise for test dataset's columns.

for i in test_data.columns:
    if(test_data[i].mean() <= (test_data[i].median() * 1.05) or test_data[i].mean() >= (test_data[i].median() * 0.95)):
        test_data[i] = test_data[i].fillna(test_data[i].mean())

    else:
        test_data[i] = test_data[i].fillna(test_data[i].median())
        
        
```

```python
# Concatinating train and test dataframes to check if there are any 0s.

data = pd.concat([train_data, test_data])

Nas = []
count_rows = data.shape[0]
for col in data.columns:
    count_nan = data[col].isna().sum() / count_rows * 100
    Nas.append(count_nan)
NAs_df = pd.DataFrame({'cols':data.columns, 'Nas_perc':Nas})
NAs_df = NAs_df.sort_values(by=['Nas_perc'], ascending=False)
NAs_df
```

```python
# Saving train and test datasets.

train_data.to_csv('../Data/TrainDataMean.csv', index=False)
test_data.to_csv('../Data/TestDataMean.csv', index=False)
```
